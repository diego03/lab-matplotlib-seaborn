# Challenge 1

#### Import all the libraries that are necessary.

```python
# import libraries here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# your code here-1st way (call `subplots` twice using the `index` parameter)
plt.plot(x,y, label = 'x-vs-y', lw=1)
plt.plot(x,z, label = 'x-vs-z', lw=3)
```

```python
# your code here-2st way (call `subplots` only once not using the `index` parameter)
fig, [ax1, ax2] = plt.subplots(1,2) #Con este parametro se indica el número de filas (1) y el número de gráficas por fila (2)
ax1.plot(x, y)
ax2.plot(x,z)
plt.show()
```

#### Augmenting your previous code, resize your previous plot.

*Hint: Add the `figsize` argument in `plt.subplots()`*

```python
# your code here
fig, [ax1, ax2] = plt.subplots(1,2, figsize=(10,5)) #Con este parametro se indica el número de filas (1) y el número de gráficas por fila (2)
ax1.plot(x, y)
ax2.plot(x,z)
plt.show()
```

#### Augmenting your previous code, label your axes.

*Hint: call `set_xlabel` and `set_ylabel`*

```python
# your code here
fig, [ax1, ax2] = plt.subplots(1,2, figsize=(10,5)) #Con este parametro se indica el número de filas (1) y el número de gráficas por fila (2)
ax1.plot(x, y)
ax1.set_title('LINEAR PLOT')
ax1.set_xlabel('X Value')
ax1.set_ylabel('Y Value')

ax2.plot(x,z)
ax2.set_title('QUADRATIC PLOT')
ax2.set_xlabel('X Value')
ax2.set_ylabel('Y Value')
plt.show()
```

#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here
x1 = np.arange(0,100)
y1 = x1**2
z1 = np.exp(x1)

plt.plot(x1,y1, label = 'x-vs-y', lw=1)
plt.plot(x1,z1, label = 'x-vs-z', lw=1)
plt.yscale('log')
plt.legend()
```

#### In the vehicles data set that you have downloaded, use the vehicles.csv file. In this exercise we will conduct some exploratory data analysis using one plot each of scatter plot, box plot, histogram, and bar chart. 

Scatter Plot

Please provide a scatter plot between "Combined MPG" as X variable and 
"Highway MPG" as Y variable

```python
# your code here
vehicles =pd.read_csv('vehicles.csv')
vehicles
```

```python
sns.lmplot(x='Combined MPG', y = 'Highway MPG', data = vehicles)
```

Box Whisker Plot

Please provide a box plot of the variable "CO2 Emission Grams/mile"

```python
#Usando un código de un línea
plt.figure(figsize=(10,6))
sns.boxplot(vehicles['CO2 Emission Grams/Mile'])
```

```python
#Otra forma más larga
vehicles_short = vehicles.drop(['Cylinders', 'Year', 'Engine Displacement', 'City MPG', 'Highway MPG', 'Combined MPG', 'Fuel Cost/Year', 'Fuel Barrels/Year'], axis = 1)

plt.figure(figsize=(10,6))
sns.boxplot(data = vehicles_short)
```

Histogram

Please provide a histogram of the Fuel Barrels/Year

```python
# your code here
#Using Pandas
vehicles['Fuel Barrels/Year'].hist(bins=100)
```

```python
#Using Seaborn
sns.distplot(vehicles['Fuel Barrels/Year'], kde= False, color='red', bins =100)
```

Bar Chart

Please provide a bar chart of the Fuel Type on the X axis and "City MPG" on the Y axis

```python
# your code here
plt.figure(figsize=(15,10))
plt.xticks(rotation =45, horizontalalignment='right')
sns.barplot(x='Fuel Type', y='City MPG', data=vehicles)
```
