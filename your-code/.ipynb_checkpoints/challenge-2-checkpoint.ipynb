## Coding Session I - 30 minutes

From the Fitbit data, we want to visually understand: 

1. **How the average number of steps change by month. Use the appropriate visualization to show the median steps by month?**
1. **Is Fitbitter more active on weekend or workdays?**

```python
# Driver 1 enter codes here
#Using Groupby
fitbit.groupby(['Months_encoded'])['Steps'].mean()
```

```python
#Using PivotTables
monthsXsteps = fitbit.pivot_table(index=['Months_encoded'], aggfunc = {'Steps':'mean'})
monthsXsteps #This assigns the result into a variable. To be used later on
```

```python
plt.figure(figsize=(15,7)) #Changes the size
sns.barplot(x=monthsXsteps.index, y='Steps', data=monthsXsteps) #The x is the index, not the name of the column since this is a pivot
plt.ylim([5000,15000])
plt.title('Average Steps x Month')
plt.xlabel('Month')
plt.ylabel('Average Steps')
```

```python
plt.figure(figsize=(15,7))
sns.lineplot(x=monthsXsteps.index, y='Steps', data=monthsXsteps)
plt.title('Average Steps x Month')
plt.xlabel('Month')
plt.ylabel('Average Steps')
```

```python
''' IF FITBITTER MORE ACTIVE ON WEEKEND OR WORKDAYS '''
#Create a PivotTable to see the results
activity = fitbit.pivot_table(index=['Days'], values = ['Steps', 'Distance', 'Calorie burned', 'Minutes Lightly Active', 'Minutes Fairly Active', 'Minutes Very Active'])

#Choose the way in which to see the table
activity[['Steps', 'Distance', 'Calorie burned', 'Minutes Lightly Active', 'Minutes Fairly Active', 'Minutes Very Active']]
```

```python
#Se asigna una tupla para que se pueda emplear el (2,3) en el tamaño
fig, ((ax1,ax2,ax3),(ax4,ax5,ax6)) = plt.subplots(2,3, figsize=(15,15)) #Con este parametro se indica el número de filas (2) y el número de gráficas por fila (3)
ax1.plot(activity.index, activity.Steps)
ax1.set_xticklabels(labels=activity.index, rotation=45,horizontalalignment='right')
ax1.set_ylabel('Average Steps')

ax2.plot(activity.index, activity.Distance, 'tab:orange')
ax2.set_xticklabels(labels=activity.index, rotation=45,horizontalalignment='right')
ax2.set_ylabel('Average Distance [kms]')

ax3.plot(activity.index, activity['Calorie burned'], 'tab:red')
ax3.set_xticklabels(labels=activity.index, rotation=45,horizontalalignment='right')
ax3.set_ylabel('Average Calories burned')

ax4.plot(activity.index, activity['Minutes Lightly Active'], 'tab:green')
ax4.set_xticklabels(labels=activity.index, rotation=45,horizontalalignment='right')
ax4.set_ylabel('Average Minutes')

ax5.plot(activity.index, activity['Minutes Fairly Active'], 'tab:grey')
ax5.set_xticklabels(labels=activity.index, rotation=45,horizontalalignment='right')
ax5.set_ylabel('Average Minutes')

ax6.plot(activity.index, activity['Minutes Very Active'],'tab:purple')
ax6.set_xticklabels(labels=activity.index, rotation=45,horizontalalignment='right')
ax6.set_ylabel('Average Minutes')
plt.show()
```

```python
''' IF FITBITTER MORE ACTIVE ON WEEKEND OR WORKDAYS '''
#Create a PivotTable to see the results
activity = fitbit.pivot_table(index=['Work_or_Weekend'], values = ['Steps', 'Distance', 'Calorie burned', 'Minutes Lightly Active', 'Minutes Fairly Active', 'Minutes Very Active'])

#Choose the way in which to see the table
activity[['Steps', 'Distance', 'Calorie burned', 'Minutes Lightly Active', 'Minutes Fairly Active', 'Minutes Very Active']]
```

```python
#Graphs
plt.figure(figsize=(10,5)) #Changes the size
sns.barplot(x=activity.index, y='Steps', data=activity) #The x is the index, not the name of the column since this is a pivot
plt.ylim([10000,11000])
plt.title('Average Steps x Type of day: Work = 1')
plt.xlabel('Work or Weekday')
plt.ylabel('Average Steps')

plt.figure(figsize=(10,5)) #Changes the size
sns.barplot(x=activity.index, y='Calorie burned', data=activity) #The x is the index, not the name of the column since this is a pivot
plt.ylim([2500,3000])
plt.title('Calories burned x Type of day: Work = 1')
plt.xlabel('Work or Weekday')
plt.ylabel('Calories burned')

plt.figure(figsize=(10,5)) #Changes the size
sns.barplot(x=activity.index, y='Minutes Very Active', data=activity) #The x is the index, not the name of the column since this is a pivot
plt.ylim([30,40])
plt.title('Minutes Very Active x Type of day: Work = 1')
plt.xlabel('Work or Weekday')
plt.ylabel('Average Minutes')
```

```python
'''There seems to be very little differences between Work/Weekdays for all of the parameters. Taking a closer look
it seems like Saturday makes up for the light activity on sundays, and thus the average amounts for week and workday are similar.

Even though the averages are similar, comparing the sums (steps, distance, calories burned), the obvious answer would be that there
is more activity during the week than during the weekend, this given the unequal amount of days being assessed (5 vs 2)'''
```

## Coding Session II - 30 minutes

**Write a loop to plot 3 scatter plots of the following features:**

1. **`Minutes Lightly Active`  vs `Steps`**
1. **`Minutes Very Active` vs `Steps`**
1. **`Minutes Sedentary` vs `Steps`**

```python
# Driver 2 enter codes here
a = sns.scatterplot(x='Minutes Lightly Active', y='Steps', data=fitbit)
b = sns.scatterplot(x = 'Minutes Very Active', y = 'Steps', data = fitbit)
c = sns.scatterplot(x = 'Minutes Sedentary', y = 'Steps', data = fitbit)
plt.show()
```

```python
fitbit2 = fitbit[['Steps', 'Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']]
fitbit2 
```

```python
eje_y='Steps'
ejes_x=['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']
```

```python
#Busque apoyo en Stackoverflow y encontré un código casí adhoc, sin embargo le quite un par de cosas
    
for x in ejes_x:
    figure = plt.figure
    ax = plt.gca()
    ax.scatter(fitbit2[x],fitbit2[eje_y])
    ax.set_xlabel(x)
    ax.set_ylabel(eje_y)
    plt.show() 
    
'''Si le quito lo último (plt.show()), se me gráfica en una sóla gráfica,sin embargo, no entiendo porqué. 
De una búsqueda en google: plt.show() starts an event loop, looks for all currently active figure objects, 
and opens one or more interactive windows that display your figure or figures.
One thing to be aware of, the plt.show() command should be used only once per Python session, 
and is most often seen at the very end of the script.'''
```
