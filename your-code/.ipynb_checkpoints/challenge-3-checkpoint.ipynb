#### Use subplots to plot the distribution of the `Age` variable with bins equal to `10`, `20` and `50`.

```python
# your code here
#Compartiendo el mismo eje "X"
fig,(ax1,ax2,ax3) = plt.subplots(1,3, figsize=(15,7), sharey=True) #Con sharex se comparte el mismo eje x
sns.distplot(titanic['Age'], kde=False, color='red', bins=10, ax=ax1)
sns.distplot(titanic['Age'], kde=False, color='red', bins=20, ax=ax2)
sns.distplot(titanic['Age'], kde=False, color='blue', bins=50, ax=ax3)
ax1.title.set_text('Age Distribution')
ax2.title.set_text('Age Distribution')
ax3.title.set_text('Age Distribution')
```

```python
#Sin compartir el mismo eje "Y"
fig,(ax1,ax2,ax3) = plt.subplots(1,3, figsize=(15,7), sharey=False) #Con sharex se comparte el mismo eje x
sns.distplot(titanic['Age'], kde=False, color='red', bins=10, ax=ax1)
sns.distplot(titanic['Age'], kde=False, color='red', bins=20, ax=ax2)
sns.distplot(titanic['Age'], kde=False, color='blue', bins=50, ax=ax3)
ax1.title.set_text('Age Distribution')
ax2.title.set_text('Age Distribution')
ax3.title.set_text('Age Distribution')
```
